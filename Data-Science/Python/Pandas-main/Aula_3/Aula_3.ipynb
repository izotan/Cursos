##**Aula 3 - Manipulando arquivos JSON**

Dados: 
- https://github.com/alura-cursos/Pandas/blob/main/pacientes.json
- https://github.com/alura-cursos/Pandas/blob/main/pacientes_2.json

###**Lendo arquivos JSON**

```python
import pandas as pd
```

```python
url = 'https://github.com/alura-cursos/Pandas/blob/main/pacientes.json?raw=True'
dados_pacientes = pd.read_json(url)
```

```python
dados_pacientes
```

```python
url = 'https://github.com/alura-cursos/Pandas/blob/main/pacientes_2.json?raw=True'
dados_pacientes_2 = pd.read_json(url)
```

```python
dados_pacientes_2
```

###**Normalizando arquivos JSON**

```python
df_normalizado = pd.json_normalize(dados_pacientes_2['Pacientes'])
```

```python
df_normalizado
```

###**Escrevendo arquivos JSON**

```python
df_normalizado.to_json('historico_pacientes_norm.json')
```

```python
pd.read_json('historico_pacientes_norm.json')
```

###**Consumindo API's utilizando leitura de arquivos JSON**

```python
#Importando bibliotecas necessárias

import requests
import json
```

```python
#Extraindo dados da api Fruityvice <https://fruityvice.com/>
dados_frutas = requests.get('https://fruityvice.com/api/fruit/all')
```

```python
#Fazendo leitura dos dados
resultado = json.loads(dados_frutas.text)
pd.DataFrame(resultado)
```

```python
#Normalizando dados para manipulação futura
resultado_normalizado = pd.json_normalize(resultado)
```
